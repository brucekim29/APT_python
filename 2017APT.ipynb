```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./dataset/서울특별시_부동산_실거래가_정보_2017년.csv',encoding='cp949')
```

```python
df.head()
```

```python
col = df.loc[ : ,['자치구명', '법정동명', '건물면적', '층정보', '건물주용도',
                  '건물명', '물건금액']].copy()
col.head()
```

```python
col.info()
```

```python
mask_1 = (col.건물주용도 =='아파트')
col_mask = col.loc[mask_1, :]
col_mask.head()
```

```python
apt_col = col_mask.copy()
apt17 = apt_col.rename(columns={'법정동명':'동이름', '건물주용도':'건물용도',
                                '건물면적':'건물면적(㎡)', '물건금액':'거래가격(만원)'})
apt17.head()
```

```python
apt17.set_index(['자치구명','동이름'],inplace=True)
apt17
```

```python
건물면적_to_평수 = 1 * 0.3025 #평수 작업 줄
apt17['평수'] = apt17['건물면적(㎡)'] * 건물면적_to_평수 #평수 column 으로 생성
apt17['평수'] = apt17['평수'].round(0)
apt17
```

```python
apt17_sorted = apt17[['건물명', '건물용도', '층정보', '거래가격(만원)', '평수', '건물면적(㎡)']]
apt17_sorted.head()
```

```python
apt17_sorted_1 = apt17_sorted.sort_values(by = '거래가격(만원)')
apt17_sorted_1
```

```python
apt17_sorted_1.to_csv('APT2017.csv')
```
