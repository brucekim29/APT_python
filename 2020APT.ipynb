```python
import pandas as pd
df = pd.read_csv('./dataset/서울특별시_부동산_실거래가_정보_2020년.csv', encoding='cp949')
df.info()
```

```python
df_1 = df.loc[:, ['자치구명','법정동명','건물면적','층정보','물건금액','건물주용도','건물명']].copy()
print(df_1)
```

```python
mask1 = (df_1.건물주용도 == "아파트")
df_apt = df_1.loc[mask1, :]
print(df_apt.head())
```

```python
df_apt2 = df_apt.rename(columns = {'물건금액':'거래가격(만원)','건물주용도' : '건물용도', '법정동명':'동이름', '건물면적': '건물면적(㎡)'})
```

```python
df_apt2
```

```python
df_apt2 = df_apt2.set_index(['자치구명', '동이름'])
print(df_apt2)
```

```python
df_apt2
```

```python
aa= 1 * 0.3025
df_apt2['평수'] = df_apt2['건물면적(㎡)'] * aa 
df_apt2
```

```python
df_apt2_sorted = df_apt2.sort_values(by = ['자치구명','동이름'])
```

```python
df_apt3 = df_apt2_sorted
df_apt3
```

```python
aa = 1 * 0.3025
```

```python
df_apt3['건물면적'] = df_apt3['건물면적(㎡)'] * aa
```
