```python
df_1 = df.loc[ : ,  ['자치구명', '법정동명', '건물면적', 
             '층정보', '건물주용도', '물건금액', '건물명']].copy()

df.head()
```

```python
print(df_1.건물주용도)
```

```python
mask1 = (df_1.건물주용도 == '아파트')
df_1_mask = df_1.loc[mask1, :]
# print(df_1_mask.head(10))
```

```python
df_1_mask.head # index 삭제하고 자치구명으로 index  설정후 sort 하면 가능
# apt.info()
```

```python
apt = df_1_mask.copy()
apt.reset_index()
apt.set_index(['자치구명', '법정동명'])
```

```python
apt1 = apt.rename(columns={'법정동명' : '동이름', '물건금액' : '거래가격(만원)',
                           '건물주용도' : '건물용도', '건물면적':'건물면적(㎡)' })
# display(apt1) #index 이름을 보기 쉽게 변경작업
```

```python
apt1.set_index(['자치구명', '동이름'], inplace=True)
# display(apt1) # 인덱스를 자치구명과 동이름으로 지정
```

```python
apt1_sorted = apt1.sort_values(by = ['자치구명','동이름'])
display(apt1_sorted) #변경한 index 에서 자치구명으로 sort 작업 진행.
```

```python
apt1_sorted = apt1_sorted[['건물명','건물용도','건물면적(㎡)','층정보','거래가격(만원)']]
display(apt1_sorted)
```

```python
건물면적_to_평수 = 1 * 0.3025 #평수 작업 줄
apt1_sorted['평수'] = apt1_sorted['건물면적(㎡)'] * 건물면적_to_평수 #평수 column 으로 만들고
print(apt1_sorted.head())
print()
apt1_sorted['평수'] = apt1_sorted['평수'].round(0).copy()
apt1_sorted
```

```python
test = apt1_sorted[['건물명', '건물용도', '층정보', '거래가격(만원)', '평수', '건물면적(㎡)']]
# test1 = test.rename(columns={'건물면적':'건물면적(㎡)'})
# test1
```

```python
test_sorted = test.sort_values(by = '거래가격(만원)')
test_sorted #변경한 index 에서 자치구명으로 sort 작업 진행.
```
